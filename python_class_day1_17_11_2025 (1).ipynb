Write a program to understand whether a number is palindrome or not

```python
num1 = input("Enter a number : ")
print(num1)
num2 = num1[: : -1]
print(num2)

if num1 == num2 :
  print(" Number is palindrome")
else :
  print(" Number is not palindrome")
```

Write a program to create a Fibnocci series

```python
n = 10
a = 0
b = 1
c = b
count = 1
print(a)
print(b)

while count <= n :
  print(c)
  count += 1
  a= b
  b=c
  c= a+b
```

Write a program to find the sum of digits of a given number

```python
num = int(input("Enter a number:"))
sum = 0
while num > 0:
  sum = sum + (num % 10 )
  num = num // 10
print(sum)
```

Write a program to reverse a number using a loop

```python
num1 = int(input("Enter a number : "))
num2 = num1
rev = 0

while num1 > 0 :
  digit = num1 % 10
  rev = rev*10 + digit
  num1 = num1 // 10
print(num2)
print(rev)
```

Write a  program to find the digits, alphabets in a string

```python
string1 = input("Enter a string :")
alphabets = 0
digits = 0
spl_char = 0

for i in range(len(string1)):
    if (string1[i].isalpha()):
       alphabets = alphabets+1
    elif (string1[i].isdigit()):
       digits = digits +1
    else :
      spl_char = spl_char +1

print("Total number of alphabets are : ", alphabets)
print("Total number of Digits are : ", digits)
print("Total number of special characters are : ", spl_char)
```
